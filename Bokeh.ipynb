```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata import iris

flowers = iris.flowers

output_file("test.html")

p = figure(title="Test Flower", x_axis_label="Petal Length", y_axis_label="Petal Width")

p.circle(
    x=flowers["petal_length"],
    y=flowers["petal_width"],
    size=8,
    color="navy",
    alpha=0.5
)

show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

output_file('test.html')

p = figure(title = "test flower")
p.xaxis.axis_label = "paltel length"
p.yaxis.axis_label = "paltel widhth"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)  # Scatter ka hi name circle hai yaha
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
x = [2,3,4,5,6,]   # this is line graph
y = [3, 4, 5, 6, 7]
output_file("line.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6,]   # this is dot graph
y = [3, 4, 5, 6, 7]
output_file("line.html")
p = figure(title = "line plot")
# this is colour change code
p.scatter(x,y,fill_color = 'red')
show(p)
```

```python
x = [2,3,4,5,6,]   # this is dot graph
y = [3, 4, 5, 6, 7]
output_file("line.html")
p = figure(title = "line plot")
# this is colour change code
p.scatter(x,y,fill_color = 'red', legend_label = 'red points')  # this is defined the label of data
show(p)
```

```python
x = [2,3,4,5,6,]   # this is dot graph
y = [3, 4, 5, 6, 7]
output_file("line.html")
p = figure(title = "line plot")
# this is colour change code
p.scatter(x,y,fill_color = 'red', legend_label = 'red points', size = 20)  # size of circle
show(p)
```
