```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.petal_length)
```

```python
sns.scatterplot(x = iris.petal_width, y = iris.petal_length)
```

```python
sns.displot(iris['sepal_length'])  # displo is equal to the histogram
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, hue = tips.smoker)
# hue ka kam hai ki ye group me data ko dikhata hai
# rel ka matlab hai ki ye relation ko batata hai
```

```python
tips.head()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data=tips, hue = 'size')

# it is represent the size
# huy colour me devide karta hai
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data=tips, style = 'size')

# style ka kam sa different different symble me devide karna
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data=tips, style = 'size', hue = 'time')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips)
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip)
```

```python
sns.pairplot(tips) #sab numerical column ko sab ke sath plot kar ke dikhata hai
```

```python
sns.pairplot(iris)
```
