catogiracle vs nomerical data

```python
x = ['a','b', 'c', 'd', 'e']  # this is catogirical data
```

```python
y = np.random.rand(5)  #this is nomiracle data
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
y = np.random.rand(5)  # horigentaly data representation
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.plot(x,y, '--y')   # --y is represent the data for yellpo colour, --b= blye colour
plt.xlabel("this is my x data")  # name likhna x, y axis ka
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
# control the figure of size
plt.figure(figsize=(6,2))
plt.plot(x,y, '--b')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()  # back side black colour
plt.show()
```
